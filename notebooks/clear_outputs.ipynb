# 🧹 Clear All Notebook Outputs
Run the single cell below to clear all outputs from all 3 notebooks.

---

```python
import json
import os
import glob

# ── Notebooks to clear ────────────────────────────────────────────────────────
NOTEBOOKS = [
    '01_data_cleaning.ipynb',
    '02_text_preprocessing.ipynb',
    '03_eda_visualizations.ipynb',
]

# ── Output images & CSVs to delete (saved in ../outputs/) ────────────────────
OUTPUTS_DIR = '../outputs/'

# ─────────────────────────────────────────────────────────────────────────────
# STEP 1: Clear notebook cell outputs
# ─────────────────────────────────────────────────────────────────────────────
def clear_notebook_outputs(filepath):
    with open(filepath, 'r', encoding='utf-8') as f:
        nb = json.load(f)
    cleared = 0
    for cell in nb['cells']:
        if cell['cell_type'] == 'code':
            cell['outputs'] = []
            cell['execution_count'] = None
            cleared += 1
    with open(filepath, 'w', encoding='utf-8') as f:
        json.dump(nb, f, indent=1, ensure_ascii=False)
    print(f'✅ Cleared {cleared} code cells  →  {filepath}')

print('=' * 55)
print('  STEP 1 — Clearing notebook cell outputs')
print('=' * 55)
for nb_name in NOTEBOOKS:
    if os.path.exists(nb_name):
        clear_notebook_outputs(nb_name)
    else:
        print(f'❌ Not found: {nb_name}  (make sure it is in the same folder)')

# ─────────────────────────────────────────────────────────────────────────────
# STEP 2: Delete all PNG images from outputs folder
# ─────────────────────────────────────────────────────────────────────────────
print()
print('=' * 55)
print('  STEP 2 — Deleting all PNG images from outputs/')
print('=' * 55)

png_files = glob.glob(OUTPUTS_DIR + '*.png')

if png_files:
    for f in png_files:
        os.remove(f)
        print(f'🗑️  Deleted  →  {f}')
else:
    print('ℹ️  No PNG files found in outputs/ folder.')

# ─────────────────────────────────────────────────────────────────────────────
# STEP 3: Delete generated CSV files from outputs folder
# ─────────────────────────────────────────────────────────────────────────────
print()
print('=' * 55)
print('  STEP 3 — Deleting generated CSVs from outputs/')
print('=' * 55)

CSV_TO_DELETE = ['data_cleaned.csv', 'data_preprocessed.csv']

for csv_name in CSV_TO_DELETE:
    csv_path = OUTPUTS_DIR + csv_name
    if os.path.exists(csv_path):
        os.remove(csv_path)
        print(f'🗑️  Deleted  →  {csv_path}')
    else:
        print(f'ℹ️  Not found (already clean): {csv_path}')

# ─────────────────────────────────────────────────────────────────────────────
# SUMMARY
# ─────────────────────────────────────────────────────────────────────────────
print()
print('=' * 55)
print(f'🎉 Done!')
print(f'   • Notebook cell outputs  : cleared')
print(f'   • PNG images deleted     : {len(png_files)}')
print(f'   • CSVs deleted           : {len(CSV_TO_DELETE)}')
print()
print('👉 You can now re-run all 3 notebooks from scratch.')
print('=' * 55)
```
